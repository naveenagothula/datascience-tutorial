```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
print(df.to_string())

g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
g=sb.FacetGrid(df,col="sex",hue="smoker")
g.map(plt.scatter,"total_bill","tip")
plt.show()

```

```python
import pandas as pd
import seaborn as sd
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
import pandas as pd
import seaborn as sd
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df)
plt.show()
```

```python
import pandas as pd
import seaborn as sd
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,hue="sex")
plt.show()
```

```python
import pandas as pd
import seaborn as sd
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,hue="size",palette="husl")
plt.show()
```

```python
import pandas as pd
import seaborn as sd
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,diag_kind='kde')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df,hue='sex',markers=["o","s"])
plt.show()
```

```python
import seaborn as sb
g=sb.pairplot(df,hue="day")
g.fig.suptitle("pairplot of tips dataset",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
